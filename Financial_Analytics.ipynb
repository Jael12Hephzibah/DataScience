```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data from the CSV file
df = pd.read_csv('/content/Financial Analytics data.csv')
```

```python
print(df.head())
```

```python
print(df.describe())
```

```python
# Correlation matrix
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
# Heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Scatter plot of Market Capitalization vs. Quarterly Sales
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Mar Cap - Crore', y='Sales Qtr - Crore', hue='Name')
plt.title('Market Capitalization vs. Quarterly Sales')
plt.xlabel('Market Capitalization (Crore)')
plt.ylabel('Quarterly Sales (Crore)')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Pairplot
sns.pairplot(df[['Mar Cap - Crore', 'Sales Qtr - Crore']])
plt.show()
```

```python
# Market Capitalization Distribution: Bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Name', y='Mar Cap - Crore', data=df, palette='viridis')
plt.title('Market Capitalization Distribution')
plt.xlabel('Company')
plt.ylabel('Market Capitalization (Crore)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Quarterly Sales Comparison: Bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Name', y='Sales Qtr - Crore', data=df, palette='muted')
plt.title('Quarterly Sales Comparison')
plt.xlabel('Company')
plt.ylabel('Quarterly Sales (Crore)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Market Capitalization vs. Quarterly Sales: Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Mar Cap - Crore', y='Sales Qtr - Crore', data=df, hue='Name', palette='colorblind')
plt.title('Market Capitalization vs. Quarterly Sales')
plt.xlabel('Market Capitalization (Crore)')
plt.ylabel('Quarterly Sales (Crore)')
plt.legend(title='Company', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
